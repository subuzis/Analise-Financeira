### IMPORTAR MÓDULOS E ARQUIVOS

```python
import pandas as pd

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#retirar coluna estado civil

funcionarios_df = funcionarios_df.drop(['Estado Civil'], axis=1)

display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

### 1- FOLHA SALARIAL

```python
#adicionando nova coluna com a soma de valores que cada funcionário recebe
funcionarios_df['Salário Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']

#pegando o resultado total
print('Total da folha salarial mensal é de R${:,}'.format(funcionarios_df['Salário Total'].sum()))
display(funcionarios_df)
```

### 2- FATURAMENTO DA EMPRESA

```python
##criando nova tabela com os faturamentos agrupando pelo ID do cliente
faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']], on='ID Cliente')

##criando nova coluna com o total de faturamento por cliente
faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']

print('O Faturamento total é de R${:,}'.format(faturamentos_df['Faturamento Total'].sum()))
display(faturamentos_df)
```

### 3- % FUNCIONÁRIOS FECHARAM CONTRATOS

```python
##unificando os valores de funcionarios da tabela serviços
qtde_func_fech_ctr_df = len(servicos_df['ID Funcionário'].unique())

##Quantidade de funcionários total
qtde_funcionarios_total_df = len(funcionarios_df['ID Funcionário'])

print('Percentual de funcionários que fecharam contrato é de {:.2%}'.format(qtde_func_fech_ctr_df / qtde_funcionarios_total_df))
```

### 4- QTDE CONTRATOS POR ÁREA

```python
##juntar tabela funcionarios com a de serviços pelo ID Funcionário
contratos_area_df = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário' , 'Area']], on='ID Funcionário')

##somando os valores
contratos_area_qtde_df = contratos_area_df['Area'].value_counts()

print(contratos_area_qtde)
contratos_area_qtde_df.plot(kind='bar')
```

### 5- FUNCIONÁRIOS POR ÁREA

```python
funcionarios_area_df = funcionarios_df['Area'].value_counts()

print(funcionarios_area_df)
funcionarios_area_df.plot(kind='bar')
```

### 6- TICKET MÉDIO MENSAL

```python
ticketm_df = clientes_df['Valor Contrato Mensal'].mean()

print('O Ticket Médio total é de R${:,.2f}'.format(ticketm_df))
```
